# 从GRIB2文件中获取要素场

使用[cfgrib](https://github.com/ecmwf/cfgrib)库加载 GRIB2 文件。

**提示**：本示例中的文件保存在 CMA-PI 高性能计算机，请在 CMA-PI 上运行或修改为本地文件路径。

从 GRAPES GFS 的原始分辨率 GRIB 2 数据中读取 850hPa 温度场。

```python
from reki.format.grib import load_field_from_file

t = load_field_from_file(
    file_path="/g1/COMMONDATA/OPER/NWPC/GRAPES_GFS_GMF/Prod-grib/2020031721/ORIG/gmf.gra.2020031800105.grb2",
    parameter="t",
    level_type="isobaricInhPa",
    level=850
)
t
```

使用 850hPa 温度场绘制填充图

```python
t.plot.contourf(size=10, aspect=2, cmap="coolwarm", levels=20)
```

省略`level`参数会加载所有层次，下面的代码加载所有等压面层的温度场。

```python
tset = load_field_from_file(
    file_path="/g1/COMMONDATA/OPER/NWPC/GRAPES_GFS_GMF/Prod-grib/2020031721/ORIG/gmf.gra.2020031800105.grb2",
    parameter="t",
    level_type="isobaricInhPa",
)
tset
```

批量绘制图形

```python
tset.plot(x="longitude", y="latitude", col="isobaricInhPa", col_wrap=3, cmap="coolwarm", levels=20)
```
